# Creating the dataset based off of adding the sasa and the categorized sasa in the sasa1.csv to the pairs_df.csv which is the faiss - global and local - code.

```python
import csv
import pandas as pd
import numpy as np
```

```python
# Load the original pairs data from a CSV file
df = pd.read_csv('/content/drive/MyDrive/Pairs_df.csv')

# Load the SASA data from another CSV file
sasa_df = pd.read_csv('/content/drive/MyDrive/sasa1.csv')
```

```python
# Add new columns for SASA values and categories; initialize as empty (None)
df['SASA_Category_A'] = None
df['SASA_Category_B'] = None
df['SASA_A'] = None
df['SASA_B'] = None
```

```python
# Debugging: Check unique identifiers before merge
unique_ids_df = df['Protein Name A'].str[:4].unique()
unique_ids_sasa_df = sasa_df['Protein_Name'].str[:4].unique()

# Convert unique identifiers to sets for set operations
set_unique_ids_df = set(unique_ids_df)
set_unique_ids_sasa_df = set(unique_ids_sasa_df)

# Find common elements
common_ids = set_unique_ids_df.intersection(set_unique_ids_sasa_df)

# Find elements to be removed from both DataFrames
deleted_from_df = list(set_unique_ids_df - common_ids)
deleted_from_sasa_df = list(set_unique_ids_sasa_df - common_ids)

# Print debugging information
print("Common IDs:", common_ids)
print("Deleted from df:", deleted_from_df)
print("Deleted from sasa_df:", deleted_from_sasa_df)

# Ensure filtering by stripping and checking exact matches
df_filtered = df[df['Protein Name A'].str[:4].str.strip().isin(common_ids)]
sasa_df_filtered = sasa_df[sasa_df['Protein_Name'].str[:4].str.strip().isin(common_ids)]

# Debugging: Check the filtered DataFrames to ensure correct filtering
print("Filtered df rows:", df_filtered)
print("Filtered sasa_df rows:", sasa_df_filtered)

# Print lengths after filtering
print("Length of unique IDs in df after filtering:", len(df_filtered['Protein Name A'].str[:4].unique()))
print("Length of unique IDs in sasa_df after filtering:", len(sasa_df_filtered['Protein_Name'].str[:4].unique()))
```

```python
# Drop rows with index 1 to 6
df = df.drop(index=range(0, 5))
```

```python
def merge_df(df, sasa_df, column_suffix):
    # Constructors for the df -> we need to modify the df since we want to do it for 2 different values
    protein_col = f'Protein Name {column_suffix}'
    sasa_col = f'SASA_{column_suffix}'
    category_col = f'SASA_Category_{column_suffix}'

    # Iterate through each row in df
    for index, row in df.iterrows():
        protein_name = row[protein_col]

        # Find matching rows in sasa_df
        matching_rows = sasa_df[sasa_df['Protein_Name'] == protein_name]

        # If a matching row is found, replace values in df
        if not matching_rows.empty:
            # Assuming you want to take the first match
            first_match = matching_rows.iloc[0]
            df.at[index, sasa_col] = first_match['SASA']
            df.at[index, category_col] = first_match['Category']

    return df

# Example usage for 'A' and 'B' columns
df = merge_df(df, sasa_df, 'A')
df = merge_df(df, sasa_df, 'B')

# Print the updated DataFrame
print(df)
```
